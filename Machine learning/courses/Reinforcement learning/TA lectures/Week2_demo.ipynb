```python
import random
import os
import math
```

```python
randomSeed = 1
random.seed(randomSeed)
```

```python
numCoins = 3
p = [0.4,0.6,0.8]
totalTosses = 20
```

# Exploration-Exploitation Dilemma

```python
totalHeads = 0
tossesFinished = 0

outcomes = []
tosses = []

for c in range (0, numCoins):
    tosses.append(0)
    o = []
    for i in range(0, totalTosses):
        o.append("-")
    outcomes.append(o)
```

```python
for i in range(0, totalTosses):

    for c in range(0, numCoins):
        s = ""
        for i in range(0, tosses[c]):
            s = s + outcomes[c][i] + " "
        print("Coin " + str(c + 1) + ": " + s)
    print("Tosses left = " + str(totalTosses - tossesFinished))
    print("------------------------------------")

    coin = input("Enter Coin (1/2/3/...) to toss: ")
    coin = int(coin)
    coin -= 1
    outcome = "t"
    if(random.uniform(0, 1.0) < p[coin]):
        outcome = "h"
        totalHeads += 1

    outcomes[coin][tosses[coin]] = outcome
    tosses[coin] += 1
    tossesFinished += 1

for c in range(0, numCoins):
    s = ""
    for i in range(0, tosses[c]):
        s = s + outcomes[c][i] + " "
    print("Coin " + str(c + 1) + ": " + s)

print("Total heads = " + str(totalHeads))
```

# Greedy with respect to mean reward

```python
totalHeads = 0
tossesFinished = 0

outcomes = []
tosses = []
means = []

for c in range (0, numCoins):
    tosses.append(0)
    means.append(0)
    o = []
    for i in range(0, totalTosses):
        o.append("-")
    outcomes.append(o)
```

```python
for i in range(0, totalTosses):

    for c in range(0, numCoins):
        s = ""
        for i in range(0, tosses[c]):
            s = s + outcomes[c][i] + " "
        print("Coin " + str(c + 1) + ": " + s)
    print("Tosses left = " + str(totalTosses - tossesFinished))
    print("Mean values:")
    print("Coin 1: "+str(means[0]))
    print("Coin 2: "+str(means[1]))
    print("Coin 3: "+str(means[2]))
    print("------------------------------------")

    coin = input("Enter Coin (1/2/3/...) to toss: ")
    coin = int(coin)
    coin -= 1
    outcome = "t"
    if(random.uniform(0, 1.0) < p[coin]):
        outcome = "h"
        totalHeads += 1

    outcomes[coin][tosses[coin]] = outcome
    tosses[coin] += 1
    tossesFinished += 1
    means[coin] = (outcomes[coin].count('h'))/tosses[coin]

for c in range(0, numCoins):
    s = ""
    for i in range(0, tosses[c]):
        s = s + outcomes[c][i] + " "
    print("Coin " + str(c + 1) + ": " + s)

print("Total heads = " + str(totalHeads))
```

# UCB 1

```python
totalHeads = 0
tossesFinished = 0

outcomes = []
tosses = []
means = []
ucb = []

for c in range (0, numCoins):
    tosses.append(0)
    means.append(0)
    ucb.append('Not defined')
    o = []
    for i in range(0, totalTosses):
        o.append("-")
    outcomes.append(o)
```

```python
for i in range(0, totalTosses):

    for c in range(0, numCoins):
        s = ""
        for i in range(0, tosses[c]):
            s = s + outcomes[c][i] + " "
        print("Coin " + str(c + 1) + ": " + s)
    print("Tosses left = " + str(totalTosses - tossesFinished))
    print("UCB values:")
    print("Coin 1: "+str(ucb[0]))
    print("Coin 2: "+str(ucb[1]))
    print("Coin 3: "+str(ucb[2]))
    print("------------------------------------")

    coin = input("Enter Coin (1/2/3/...) to toss: ")
    coin = int(coin)
    coin -= 1
    outcome = "t"
    if(random.uniform(0, 1.0) < p[coin]):
        outcome = "h"
        totalHeads += 1

    outcomes[coin][tosses[coin]] = outcome
    tosses[coin] += 1
    tossesFinished += 1
    means[coin] = (outcomes[coin].count('h'))/tosses[coin]
    for cn in range(0, numCoins):
      if (tosses[cn] != 0):
        ucb[cn] = means[cn] + math.sqrt((2*math.log(tossesFinished))/tosses[cn])

for c in range(0, numCoins):
    s = ""
    for i in range(0, tosses[c]):
        s = s + outcomes[c][i] + " "
    print("Coin " + str(c + 1) + ": " + s)

print("Total heads = " + str(totalHeads))
```
