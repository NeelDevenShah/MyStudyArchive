```python
import random
import os
import math
import numpy as np
```

```python
randomSeed = 1
random.seed(randomSeed)
```

```python
numCoins = 3
p = [0.2,0.5,0.8]
totalTosses = 30
```

# Exploration-Exploitation Dilemma

```python
totalHeads = 0
tossesFinished = 0

outcomes = []
tosses = []

for c in range (0, numCoins):
    tosses.append(0)
    o = []
    for i in range(0, totalTosses):
        o.append("-")
    outcomes.append(o)
```

```python
for i in range(0, totalTosses):

    for c in range(0, numCoins):
        s = ""
        for i in range(0, tosses[c]):
            s = s + outcomes[c][i] + " "
        print("Coin " + str(c + 1) + ": " + s)
    print("Tosses left = " + str(totalTosses - tossesFinished))
    print("------------------------------------")

    coin = input("Enter Coin (1/2/3/...) to toss: ")
    coin = int(coin)
    coin -= 1
    outcome = "t"
    if(random.uniform(0, 1.0) < p[coin]):
        outcome = "h"
        totalHeads += 1

    outcomes[coin][tosses[coin]] = outcome
    tosses[coin] += 1
    tossesFinished += 1

for c in range(0, numCoins):
    s = ""
    for i in range(0, tosses[c]):
        s = s + outcomes[c][i] + " "
    print("Coin " + str(c + 1) + ": " + s)

print("Total heads = " + str(totalHeads))
```

# Greedy with respect to mean reward

```python
totalHeads = 0
tossesFinished = 0

outcomes = []
tosses = []
means = []

for c in range (0, numCoins):
    tosses.append(0)
    means.append(0)
    o = []
    for i in range(0, totalTosses):
        o.append("-")
    outcomes.append(o)
```

```python
for i in range(0, totalTosses):

    for c in range(0, numCoins):
        s = ""
        for i in range(0, tosses[c]):
            s = s + outcomes[c][i] + " "
        print("Coin " + str(c + 1) + ": " + s)
    print("Tosses left = " + str(totalTosses - tossesFinished))
    print("Mean values:")
    print("Coin 1: "+str(means[0]))
    print("Coin 2: "+str(means[1]))
    print("Coin 3: "+str(means[2]))
    print("------------------------------------")

    coin = input("Enter Coin (1/2/3/...) to toss: ")
    coin = int(coin)
    coin -= 1
    outcome = "t"
    if(random.uniform(0, 1.0) < p[coin]):
        outcome = "h"
        totalHeads += 1

    outcomes[coin][tosses[coin]] = outcome
    tosses[coin] += 1
    tossesFinished += 1
    means[coin] = (outcomes[coin].count('h'))/tosses[coin]

for c in range(0, numCoins):
    s = ""
    for i in range(0, tosses[c]):
        s = s + outcomes[c][i] + " "
    print("Coin " + str(c + 1) + ": " + s)

print("Total heads = " + str(totalHeads))
```

# UCB 1

```python
totalHeads = 0
tossesFinished = 0

outcomes = []
tosses = []
means = []
ucb = []

for c in range (0, numCoins):
    tosses.append(0)
    means.append(0)
    ucb.append('Not defined')
    o = []
    for i in range(0, totalTosses):
        o.append("-")
    outcomes.append(o)
```

```python
for i in range(0, totalTosses):

    for c in range(0, numCoins):
        s = ""
        for i in range(0, tosses[c]):
            s = s + outcomes[c][i] + " "
        print("Coin " + str(c + 1) + ": " + s)
    print("Tosses left = " + str(totalTosses - tossesFinished))
    print("UCB values:")
    print("Coin 1: "+str(ucb[0]))
    print("Coin 2: "+str(ucb[1]))
    print("Coin 3: "+str(ucb[2]))
    print("------------------------------------")

    coin = input("Enter Coin (1/2/3/...) to toss: ")
    coin = int(coin)
    coin -= 1
    outcome = "t"
    if(random.uniform(0, 1.0) < p[coin]):
        outcome = "h"
        totalHeads += 1

    outcomes[coin][tosses[coin]] = outcome
    tosses[coin] += 1
    tossesFinished += 1
    means[coin] = (outcomes[coin].count('h'))/tosses[coin]
    for cn in range(0, numCoins):
      if (tosses[cn] != 0):
        ucb[cn] = means[cn] + math.sqrt((2*math.log(tossesFinished))/tosses[cn])

for c in range(0, numCoins):
    s = ""
    for i in range(0, tosses[c]):
        s = s + outcomes[c][i] + " "
    print("Coin " + str(c + 1) + ": " + s)

print("Total heads = " + str(totalHeads))
```

# Thomson Sampling

```python
totalHeads = 0
tossesFinished = 0

outcomes = []
tosses = []
successes = []
failures = []
betaValues = []

for c in range (0, numCoins):
    tosses.append(0)
    successes.append(0)
    failures.append(0)
    betaValues.append(np.random.beta(0+1, 0+1))
    o = []
    for i in range(0, totalTosses):
        o.append("-")
    outcomes.append(o)
```

```python
for i in range(0, totalTosses):

    for c in range(0, numCoins):
        s = ""
        for i in range(0, tosses[c]):
            s = s + outcomes[c][i] + " "
        print("Coin " + str(c + 1) + ": " + s)
    print("Tosses left = " + str(totalTosses - tossesFinished))
    print("Beta values:")
    print("Coin 1: "+str(betaValues[0]))
    print("Coin 2: "+str(betaValues[1]))
    print("Coin 3: "+str(betaValues[2]))
    print("------------------------------------")

    coin = input("Enter Coin (1/2/3/...) to toss: ")
    coin = int(coin)
    coin -= 1
    outcome = "t"
    if(random.uniform(0, 1.0) < p[coin]):
        outcome = "h"
        totalHeads += 1

    outcomes[coin][tosses[coin]] = outcome
    tosses[coin] += 1
    tossesFinished += 1
    successes[coin] = outcomes[coin].count('h')
    failures[coin] = outcomes[coin].count('t')
    for cn in range(0, numCoins):
      betaValues[cn] = np.random.beta(successes[cn]+1, failures[cn]+1)

for c in range(0, numCoins):
    s = ""
    for i in range(0, tosses[c]):
        s = s + outcomes[c][i] + " "
    print("Coin " + str(c + 1) + ": " + s)

print("Total heads = " + str(totalHeads))
```
