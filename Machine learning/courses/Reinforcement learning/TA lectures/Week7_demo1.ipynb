```python
numCoins = 3
p = [0.2,0.5,0.8]
totalTosses = 30
```

# Maximum Likelihood Estimate (MLE)

```python
totalHeads = 0
tossesFinished = 0

outcomes = []
tosses = [] # Number of tosses of each coin
means = [] # Mean of each coin

for c in range (0, numCoins):
    tosses.append(0)
    means.append(0)
    o = []
    for i in range(0, totalTosses):
        o.append("-")
    outcomes.append(o)
```

Explore each coin initially otherwise might end up picking just one coin all the time

```python
for i in range(0, totalTosses):

    for c in range(0, numCoins):
        s = ""
        for i in range(0, tosses[c]):
            s = s + outcomes[c][i] + " "
        print("Coin " + str(c + 1) + ": " + s)
    print("Tosses left = " + str(totalTosses - tossesFinished))
    print("Mean values:")
    print("Coin 1: "+str(means[0]))
    print("Coin 2: "+str(means[1]))
    print("Coin 3: "+str(means[2]))
    print("------------------------------------")

    coin = input("Enter Coin (1/2/3/...) to toss: ")
    coin = int(coin)
    coin -= 1
    outcome = "t"
    if(random.uniform(0, 1.0) < p[coin]):
        outcome = "h"
        totalHeads += 1

    outcomes[coin][tosses[coin]] = outcome
    tosses[coin] += 1
    tossesFinished += 1
    means[coin] = (outcomes[coin].count('h'))/tosses[coin]

for c in range(0, numCoins):
    s = ""
    for i in range(0, tosses[c]):
        s = s + outcomes[c][i] + " "
    print("Coin " + str(c + 1) + ": " + s)

print("Total heads = " + str(totalHeads))
```

# Maximum A Posteriori (MAP) Estimate

```python
totalHeads = 0
tossesFinished = 0

outcomes = []
tosses = [] # Number of tosses of each coin
successes = [] # Number of successes of each coin (heads)
failures = [] # Number of failures of each coin (tails)
meanValues = [] # Mean of each coin

for c in range (0, numCoins):
    tosses.append(0)
    successes.append(0)
    failures.append(0)
    meanValues.append(np.random.beta(0+1, 0+1))
    o = []
    for i in range(0, totalTosses):
        o.append("-")
    outcomes.append(o)
```

+1 for 3 reasons:



1. Avoid stuffs like 0/0 which cause implementation issues
2. Fairness: 0.5 for unsampled coin
3. Beta(1,1) is uniform distribution and Beta(0,0) is not defined

As we encounter large number of successes and failures, +1 should not matter



```python
for i in range(0, totalTosses):

    for c in range(0, numCoins):
        s = ""
        for i in range(0, tosses[c]):
            s = s + outcomes[c][i] + " "
        print("Coin " + str(c + 1) + ": " + s)
    print("Tosses left = " + str(totalTosses - tossesFinished))
    print("Mean values:")
    print("Coin 1: "+str(meanValues[0]))
    print("Coin 2: "+str(meanValues[1]))
    print("Coin 3: "+str(meanValues[2]))
    print("------------------------------------")

    coin = input("Enter Coin (1/2/3/...) to toss: ")
    coin = int(coin)
    coin -= 1
    outcome = "t"
    if(random.uniform(0, 1.0) < p[coin]):
        outcome = "h"
        totalHeads += 1

    outcomes[coin][tosses[coin]] = outcome
    tosses[coin] += 1
    tossesFinished += 1
    successes[coin] = outcomes[coin].count('h')
    failures[coin] = outcomes[coin].count('t')
    for cn in range(0, numCoins):
      meanValues[cn] = (successes[cn]+1)/((successes[cn]+1)+(failures[cn]+1)) # RHS is mean of Beta distribution

for c in range(0, numCoins):
    s = ""
    for i in range(0, tosses[c]):
        s = s + outcomes[c][i] + " "
    print("Coin " + str(c + 1) + ": " + s)

print("Total heads = " + str(totalHeads))
```

# Full Bayesian: Thomson Sampling

```python
totalHeads = 0
tossesFinished = 0

outcomes = []
tosses = [] # Number of tosses of each coin
successes = [] # Number of successes of each coin (heads)
failures = [] # Number of failures of each coin (tails)
betaValues = [] # Mean of each coin

for c in range (0, numCoins):
    tosses.append(0)
    successes.append(0)
    failures.append(0)
    betaValues.append(np.random.beta(0+1, 0+1))
    o = []
    for i in range(0, totalTosses):
        o.append("-")
    outcomes.append(o)
```

Implicitly enforces exploration

```python
for i in range(0, totalTosses):

    for c in range(0, numCoins):
        s = ""
        for i in range(0, tosses[c]):
            s = s + outcomes[c][i] + " "
        print("Coin " + str(c + 1) + ": " + s)
    print("Tosses left = " + str(totalTosses - tossesFinished))
    print("Beta values:")
    print("Coin 1: "+str(betaValues[0]))
    print("Coin 2: "+str(betaValues[1]))
    print("Coin 3: "+str(betaValues[2]))
    print("------------------------------------")

    coin = input("Enter Coin (1/2/3/...) to toss: ")
    coin = int(coin)
    coin -= 1
    outcome = "t"
    if(random.uniform(0, 1.0) < p[coin]):
        outcome = "h"
        totalHeads += 1

    outcomes[coin][tosses[coin]] = outcome
    tosses[coin] += 1
    tossesFinished += 1
    successes[coin] = outcomes[coin].count('h')
    failures[coin] = outcomes[coin].count('t')
    for cn in range(0, numCoins):
      betaValues[cn] = np.random.beta(successes[cn]+1, failures[cn]+1)

for c in range(0, numCoins):
    s = ""
    for i in range(0, tosses[c]):
        s = s + outcomes[c][i] + " "
    print("Coin " + str(c + 1) + ": " + s)

print("Total heads = " + str(totalHeads))
```
